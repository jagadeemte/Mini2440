## Visualizations

Visualization is a key factor in understanding deep learning models and their behavior. Typically, pyplot from the matplotlib package is used, capable of visualizing series and 2D data.

Below is an example of visualizing series data.

```python
x = np.linspace(-5, 5, 50) # create a linear spacing from x = -5.0 to 5.0 with 50 steps

y1 = x**2      # create a series of points {y1}, which corresponds to the function f(x) = y^2
y2 = 4*np.sin(x) # create another series of points {y2}, which corresponds to the function f(x) = 4*sin(x)  NOTE: we have to use np.sin and not math.sin as math.sin will only act on individual values
# to use math.sin, we could have used a list comprehension instead: y2 = [math.sin(xi) for xi in x]

# by default, matplotlib will behave like MATLAB with hold(True), overplotting until a new figure object is created
plt.plot(x, y1, label="x^2")        # plot y1 with x as the x-axis series, and label the line "x^2"
plt.plot(x, y2, label="4 sin(x)")   # plot y2 with x as the x-axis series, and label the line "4 sin(x)"
plt.legend()                        # have matplotlib show the label on the plot
```

```python
from scipy.special import erf  # Import the error function

x = np.linspace(-4, 4, 50) # create a linear spacing from x = -4.0 to 4.0 with 50 steps

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the Hyperbolic Tangent function
def hyperbolictangent(x):
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))

def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))

# Define the Tanh GELU function
def tanh_gelu(x):
    return 0.5 * x * (1 + np.tanh(x/np.sqrt(2)))

# create and plot the functions below
# Computing y values using NumPy's built-in maximum function for ReLU
y1 = np.maximum(0, x)
y2 = np.maximum(0.1*x,x)
# Compute y values using the sigmoid function
y3 = sigmoid(x)
/#Hyperbolic Tangent:  tanh(x)=(ex−e−x)/(ex+e−x)
y4 = hyperbolictangent(x)
#SiLu
y5 = x * sigmoid(x)
#GeLu function
y6 = gelu(x)

y7 = tanh_gelu(x)

# Plot the ReLU function
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='ReLU(x) = max(x, 0)', color='blue', linewidth=2)
plt.plot(x, y2, label='Leaky-ReLu', color='red', linewidth=2)
plt.plot(x, y3, label='Sigmoid', color='green', linewidth=2)
plt.plot(x, y4, label = 'Hyperbolic Tangent', color='orange', linewidth=2)
plt.plot(x, y5, label = 'SiLU',color='yellow',linewidth=2)
plt.plot(x, y6, label = 'Gelu' , color = 'violet', linewidth = 2)
plt.plot(x, y7, label = 'TanhGelu', color ='black', linewidth = 2)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Add horizontal line
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Add vertical line
#plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("Fn(x)")
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

### Visualizing 2D data

In many cases, we also want the ability to visualize multi-dimensional data such as images. To do so, matplotlib has the imshow method, which can visualize single channel data with a heatmap, or RGB data with color.

Let's consider visualizing the first 8 training images from the MNIST dataset. MNIST consists of hand drawn digits with their corresponding labels (a number from 0 to 9).

We will use the tensorflow keras dataset library to load the dataset, and then visualize the images with a matplotlib subplot.
Because we have so many images, we should arrange them in a grid (4 horizontal, 2 vertical), and plot each image in a loop.
Furthermore, we can append the label to each image using the matplotlib utility.


```python
from tensorflow.keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Define the grid dimensions
rows, cols = 2, 4

# Create a figure and axes for the grid
fig, axes = plt.subplots(rows, cols, figsize=(8, 5))

# Iterate through the grid
for i in range(rows):
    for j in range(cols):
        index = i * cols + j
        ax = axes[i, j]

        # Display the image
        ax.imshow(train_images[index], cmap='gray')

        # Display the label on top of the image in red text
        ax.text(0.9, 0.9, str(train_labels[index]), color='red',
                transform=ax.transAxes, fontsize=24,
                ha='center', va='center')

        # Turn off axis labels
        ax.axis('off')

# Adjust spacing and layout
plt.tight_layout()
```

### Visualizing Tensors

Aside from visualzing linear functions and images, we can also visualize entire tensors from DL models.

```python
# first, let's download an existing model to inspect
model = tf.keras.applications.VGG16(weights='imagenet')

# can then print the summary of what the model is composed of
print(model.summary())
```

```python
# we can also print the model layers based on index to better understand the structure
for i,layer in enumerate(model.layers):
  print(f"{i}: {layer}")
```

Not all of these layers contain weights, for example, MaxPooling2D is a stateless operation, and so is Flatten.
Conv2D and Dense are the two layer types that can be visualized.
That said, let's visualize the filter kernels in the first convoluton layer.

```python
# next we can extract som
layer = model.layers[1] # Get the first convolutional layer
weights = layer.get_weights()[0]

n_filters = weights.shape[-1]

for i in range(n_filters):
    plt.subplot(8, 8, i+1)  # Assuming 64 filters, adjust if necessary
    plt.imshow(weights[:, :, 0, i], cmap="viridis")
    plt.axis('off')
```

Aside from visualizing the weights directly, we can also compute and visualize the weight distribution using a histogram.

```python

# we can use the mean and var (variance) functions built in to calculate some simple statistics
print(f"weight tensor has mean: {weights.mean()} and variance: {weights.var()}")

# we need to call .flatten() on the tensor so that all the histogram sees them as a 1D array. Then we can plot with 100 bins to get a bit more resolution in the histogram.
plt.hist(weights.flatten(), bins=100)
```

When analyzing the activation distributions and comparing them to the weight tensor distributions, several key insights can be drawn. Let’s break it down:

What to Notice About Activation Distributions
Range of Activation Values:

ReLU Layers (e.g., fc1, fc2):
Activations are non-negative because ReLU clamps negative values to zero.
Distributions may have a peak at zero (sparse activations) if many neurons are inactive for the input.
Softmax Layer (predictions):
Outputs represent probabilities, so values are between 0 and 1.
The distribution may have one or a few high values (for likely classes) and many low values (for unlikely classes).
Sparsity:

For dense layers with ReLU, sparsity indicates that only a subset of neurons are actively contributing, which helps the network avoid overfitting.
Variability Across Layers:

Activations in earlier dense layers (fc1, fc2) typically have wider distributions, as they capture intermediate features.
The final layer (predictions) usually has concentrated values because it is designed to output probabilities for specific classes.
Shape of Distributions:

Look for normal-like distributions in fc1 and fc2, which may indicate effective weight initialization.
Unusual patterns, like a large spike at zero or extreme skewness, might suggest issues like vanishing gradients or overly sparse activations.
Comparison to Weight Tensor Distributions
Weight Tensor Distributions:

Typically initialized using strategies like Xavier or He initialization, which aim for a balanced range (e.g., Gaussian or uniform distributions).
The weights are not sparse but spread across a range to ensure stable forward and backward propagation.
Key Differences:

Weights:
Represent learnable parameters. Their distribution remains roughly consistent across layers if initialization is done correctly.
Activations:
Represent the outputs of applying weights and activation functions to the input. These are highly dependent on the input data and vary across layers.
Patterns to Observe:

If the weight distribution shows values concentrated near zero, activations may also become sparse, especially with ReLU.
Overly broad weight distributions can lead to extreme activations (e.g., exploding gradients), affecting training stability.
A mismatch between the two distributions (e.g., extremely sparse activations for well-distributed weights) may indicate issues with the activation function or model design.
